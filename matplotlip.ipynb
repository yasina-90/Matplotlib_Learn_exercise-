# Matplotlib for beginners


Matplotlib is a library for making 2D plots in Python. It is
designed with the philosophy that you should be able to
create simple plots with just a few commands:


1 Initialize

```python
import numpy as np
import matplotlib.pyplot as plt
```

2 Prepare


```python
X = np.linspace(0, 10*np.pi, 1000)
Y = np.sin(X)
```

3 Render and Observe

```python
fig, ax = plt.subplots()
ax.plot(X, Y)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data for line plot and scatter plot
x = np.linspace(0, 10, 100)  # Array of 100 evenly spaced points from 0 to 10
y = np.sin(x)                 # Sine values of x

# Sample data for bar plot and histogram
data = np.random.normal(loc=0, scale=1, size=1000)  # Generate 1000 random numbers from a normal distribution
```

```python
plt.figure()  # Create a new figure
plt.plot(x, y)  # Plot x versus y
plt.title('Line Plot, Sine Function')  # Add a title
plt.xlabel('X-axis')    # Label for x-axis
plt.ylabel('Y-axis')    # Label for y-axis
plt.grid(True)          # Show grid
plt.show()              # Show the plot
```

```python
plt.figure()  # Create a new figure
plt.scatter(x, y, c='r', marker='o', label='Data')  # Scatter plot of x versus y
plt.title('Scatter Plot')  # Add a title
plt.xlabel('X-axis')        # Label for x-axis
plt.ylabel('Y-axis')        # Label for y-axis
plt.legend()                # Show legend
plt.grid(True)              # Show grid
plt.show()                  # Show the plot
```

```python
plt.figure()  # Create a new figure
plt.hist(data, bins=30, color='skyblue', edgecolor='black')  # Plot histogram of data with 30 bins
plt.title('Histogram')  # Add a title
plt.xlabel('Value')     # Label for x-axis
plt.ylabel('Frequency') # Label for y-axis
plt.grid(True)          # Show grid
plt.show()              # Show the plot
```

```python
plt.figure()  # Create a new figure
plt.bar(np.arange(len(data)), data, color='orange')  # Plot bar chart of data
plt.title('Bar Plot')    # Add a title
plt.xlabel('Index')      # Label for x-axis
plt.ylabel('Value')      # Label for y-axis
plt.grid(True)           # Show grid
plt.show()               # Show the plot
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Create a figure with 2 subplots arranged vertically
fig, axs = plt.subplots(2, 1, figsize=(8, 6))  # 2 rows, 1 column, figsize is the size of the entire figure

# Plot data on the first subplot
axs[0].plot(x, y1, 'r-', label='sin(x)')  # Plot sin(x) with a red solid line
axs[0].set_title('Sin Plot')  # Set title for the first subplot
axs[0].legend()  # Show legend on the first subplot
# Plot data on the second subplot
axs[1].plot(x, y2, 'b--', label='cos(x)')  # Plot cos(x) with a blue dashed line
axs[1].set_title('Cos Plot')  # Set title for the second subplot
axs[1].legend()  # Show legend on the second subplot

# Adjust layout to prevent overlap of subplots
plt.tight_layout()

# Show the plots

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create a plot
plt.plot(x, y, color='green', linestyle='--', linewidth=2, marker='o', markersize=5, label='Sin(x)')
# 'green' color, dashed line style, line width of 2, circular markers with size 5

# Add labels and title
plt.xlabel('X-axis')  # Label for x-axis
plt.ylabel('Y-axis')  # Label for y-axis
plt.title('Customized Plot')  # Title of the plot

# Add grid
plt.grid(True)  # Show grid on the plot

# Add legend
plt.legend()  # Show legend on the plot

# Show the plot
plt.show()
```

### Colors:

-   `b`: Blue
-   `g`: Green
-   `r`: Red
-   `c`: Cyan
-   `m`: Magenta
-   `y`: Yellow
-   `k`: Black
-   `w`: White

### Line Styles:

-   `-`: Solid line
-   `--`: Dashed line
-   `-.`: Dash-dot line
-   `:`: Dotted line

### Markers (for scatter plots and line plots with markers):

-   `.`: Point marker
-   `,`: Pixel marker
-   `o`: Circle marker
-   `s`: Square marker
-   `+`: Plus marker
-   `x`: X marker
-   `*`: Star marker
-   `D`: Diamond marker
-   `^`: Triangle up marker
-   `v`: Triangle down marker
-   `>`: Triangle right marker
-   `<`: Triangle left marker
-   `p`: Pentagram marker
-   `h`: Hexagon marker


```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.linspace(0, 10, 100)
y1 = np.sin(x) * 1.5  # Scale the sine function's y-values
y2 = np.cos(x) * 1.5  # Scale the cosine function's y-values
y3 = np.tan(x)

# Create a plot with multiple lines and markers
plt.plot(x, y1, 'r--', label='1.5*sin(x)')  # Red dashed line for scaled sine function
plt.plot(x, y2, 'g-.', label='1.5*cos(x)')  # Green dash-dot line for scaled cosine function
plt.plot(x, y3, 'b:', label='tan(x)')   # Blue dotted line for tangent function

# Add markers to specific points
plt.plot([0, 2, 4, 6, 8, 10], [0, 1.5, -1.5, 1.5, -1.5, 1.5], 'ko', label='Points')  # Black circles for specific points

# Add labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Example Plot')

# Add legend
plt.legend()

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.linspace(0, 10, 100)
y1 = np.sin(x) * 1.5
y2 = np.cos(x) * 1.5

# Create a figure with two subplots arranged horizontally
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))  # 1 row, 2 columns

# Plot data on the first subplot
ax1.plot(x, y1, color='red')
ax1.set_title('Sine Function')
ax1.set_xlabel('X-axis')
ax1.set_ylabel('Y-axis')
ax1.grid(True)

# Plot data on the second subplot
ax2.plot(x, y2, color='blue')
ax2.set_title('Cosine Function')
ax2.set_xlabel('X-axis')
ax2.set_ylabel('Y-axis')
ax2.grid(True)

# Adjust layout
plt.tight_layout()

# Add annotations with arrows
ax1.annotate('Peak', xy=(np.pi/2, 1.5), xytext=(np.pi/2+1, 2),
             arrowprops=dict(facecolor='black', arrowstyle='->'))
ax2.annotate('Peak', xy=(0, 1.5), xytext=(1, 2),
             arrowprops=dict(facecolor='black', arrowstyle='->'))

# Show the plots
plt.show()
```

```python
# Plot with customized appearance
plt.plot(x, y1, color='red', linestyle='--', linewidth=2, marker='o', markersize=8, label='Sine Function')
plt.plot(x, y2, color='blue', linestyle='-.', linewidth=2, marker='s', markersize=8, label='Cosine Function')

# Add legend with customized location and font size
plt.legend(loc='upper right', fontsize=10)

# Customize labels, title, and grid
plt.xlabel('X-axis', fontsize=12)
plt.ylabel('Y-axis', fontsize=12)
plt.title('Customized Plot', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.5)

# Show the plot
plt.show()
```

```python
# Plot with logarithmic y-axis
plt.plot(x, y1)
plt.yscale('log')

# Add labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis (log scale)')
plt.title('Logarithmic Scale')

# Show the plot
plt.show()
```

```python
# Sample data for contour plot
x = np.linspace(-3, 3, 100)
y = np.linspace(-3, 3, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(X**2 + Y**2)

# Create a contour plot
plt.contour(X, Y, Z, cmap='viridis')

# Add colorbar and title
plt.colorbar()
plt.title('Contour Plot')

# Show the plot
plt.show()
```

```python
# Sample data for streamplot
Y, X = np.mgrid[-3:3:100j, -3:3:100j]
U = -1 - X**2 + Y
V = 1 + X - Y**2

# Create a streamplot
plt.streamplot(X, Y, U, V, color='skyblue')

# Add title
plt.title('Streamplot')

# Show the plot
plt.show()
```

```python
# Sample data for quiver plot
X, Y = np.meshgrid(np.arange(-2, 2, 0.5), np.arange(-2, 2, 0.5))
U = -1 - X**2 + Y
V = 1 + X - Y**2

# Create a quiver plot
plt.quiver(X, Y, U, V)

# Add title
plt.title('Quiver Plot')

# Show the plot
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

# Sample data for 3D surface plot
X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y)
Z = np.sin(np.sqrt(X**2 + Y**2))

# Create a 3D surface plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')

# Add title
ax.set_title('3D Surface Plot')

# Show the plot
plt.show()
```
